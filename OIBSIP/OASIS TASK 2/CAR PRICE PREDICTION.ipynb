```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {'n_estimators': [100,200,300,400,500,600,700,800,900,1000],
             'criterion': ['squared_error', 'absolute_error', 'poisson', 'friedma_mse'],
             'max_depth': [10,20,30,40,50],
             'min_samples_split': [2,5,10,20,50],
             'min_samples_leaf': [1,2,5,10],
             'max_features': ['auto', 'sqrt', 'log2']
             }
```

```python
parameters
```

```python
random_cv = RandomizedSearchCV(estimator=regressor, param_distributions=parameters, n_iter=10, scoring='neg_mean_absolute_error',
                              random_state=42, cv = 5, verbose = 2, n_jobs=-1) 
```

```python
random_cv.fit(x_train, y_train)
```

```python
random_cv.best_params_
```

```python
random_cv.best_score_
```

```python
predictions = random_cv.predict(x_test)
```

```python
sns.displot(y_test-predictions)
```

```python
plt.scatter(y_test, predictions)
```

```python
from sklearn import metrics
```

```python
print('MAE:',metrics.mean_absolute_error(y_test, predictions))
print('MAE:',metrics.mean_squared_error(y_test, predictions))
print('MAE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
